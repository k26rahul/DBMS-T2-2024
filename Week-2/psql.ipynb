### psql (PostgreSQL interactive terminal)

ðŸ”— Documentation (https://www.postgresql.org/docs/current/app-psql.htm)

- **psql**:
  - Terminal-based front-end to PostgreSQL.
  - Allows interactive SQL query input and execution.

- **Prompt**:
  - Displays current database name followed by `=#`.

- **SQL Commands**:
  - Type SQL commands at the prompt.
  - Commands are sent to the server when a semicolon `;` is reached.
  - Commands can span multiple lines.

- **Meta-Commands**:
  - Begin with an unquoted backslash `\`.
  - Processed by psql itself, useful for administration and scripting.
  - Known as slash or backslash commands.

### psql Commands

- **Login (default db)**:
  ```sh
  psql -U username
  ```

- **Login (to a db)**:
  ```sh
  psql -U username -d dbname
  ```

- **Connection info**:
  ```sh
  \c
  \conninfo
  ```

- **List all db**:
  ```sh
  \l
  ```

- **Connect to a db**:
  ```sh
  \c dbname
  ```

- **List all schemas**:
  ```sh
  \dn
  \dn *
  ```

- **List all tables**:
  ```sh
  \dt
  ```

- **Describe table**:
  ```sh
  \d table
  ```

- **List all views**:
  ```sh
  \dv
  ```

- **List all functions**:
  ```sh
  \df
  ```

- **List all users**:
  ```sh
  \du
  ```

- **Execute SQL file**:
  ```sh
  \i file.sql
  ```

- **Clear screen**:
  ```sh
  \! cls
  ```

- **Quit**:
  ```sh
  \q
  ```
